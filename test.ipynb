```python
import json
import pandas as pd
```

```python
with open('data/data.json', 'r') as file:
    data = json.load(file)

print(data)
```

```python
df = pd.DataFrame(
    [(user, movie, rating) for user, movies in data.items() for movie, rating in movies.items()],
    columns=["user", "movie", "rating"]
)
```
